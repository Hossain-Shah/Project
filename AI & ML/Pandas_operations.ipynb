```python
#dataframe, series create and csv conversion

# Import pandas package
import pandas as pd

# Define a dictionary containing employee data
data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
series = pd.Series(data)
print(series[['Name', 'Qualification']])
df.to_csv ('/content/drive/MyDrive/Colab_Notebooks/test.csv', index='Name', header = 'False')

# select two columns
print(df[['Name', 'Qualification']])
```

```python
#group sort

group = ['A', 'B', 'C', 'C']
df['Group'] = group
average_age = df['Age'].groupby(df['Group'])
print(average_age.mean())
```

```python
#concat string with delimiter

df["ID"] = df['Name'].astype(str) + ''.join('_') + df['Qualification'] +''.join('-') + df['Address']
print(df)
```

```python
#data loop-wise lock

print(df.iloc[1:3])
```

```python
#dropping table

df.dropna()
```

```python
#conditional banding

import numpy as np

df.loc[df['Age'] < 25, 'Age'] = np.nan
print(df)
df.dropna()
```

```python
#index-wise iteration

data = pd.Series([((n*(n+1))/2) for n in range(7)], index = [i for i in range(7)])
print(data)

data = pd.DataFrame([((n*(n+1))/2) for n in range(7)], index = [i for i in range(7)])
print(data)
```

```python
#range-wise bounded iteration

# import pandas and numpy
import pandas as pd
import numpy as np

# series with numpy linspace()
ser1 = pd.Series(np.linspace(3, 33, 3))
print(ser1)

df = pd.DataFrame(np.linspace(3, 33, 3))
print(df)
```
